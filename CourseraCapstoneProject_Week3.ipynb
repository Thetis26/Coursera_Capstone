# Part 2

I added two columns for latitude and longitude and read the cvs containing the latitude and longitude of each postal code

```python
df.insert(3,"Latitude",None,True)
df.insert(4,"Longitude",None,True)
postal_codes_cvs = pd.read_csv('Geospatial_Coordinates.csv') 
postal_codes_cvs
```

I looped through each postal code in the neighborhood dataframe and i assigned the coresponding latitude and longitude

```python
# loop until you get the coordinates
for postal_code in df.PostalCode:
    lat = postal_codes_cvs.loc[postal_codes_cvs['Postal Code'] == postal_code,'Latitude']
    lng = postal_codes_cvs.loc[postal_codes_cvs['Postal Code'] == postal_code,'Longitude']
    lat = float(str(lat).split()[1])
    lng = float(str(lng).split()[1])
    df.loc[df['PostalCode'] == postal_code,'Latitude'] = lat
    df.loc[df['PostalCode'] == postal_code,'Longitude'] = lng
df.head()
```

```python
#!pip3 install geopy
from geopy.geocoders import Nominatim # convert an address into latitude and longitude values

import requests # library to handle requests
from pandas.io.json import json_normalize # tranform JSON file into a pandas dataframe

# Matplotlib and associated plotting modules
import matplotlib.cm as cm
import matplotlib.colors as colors

# import k-means from clustering stage
#!pip3 install -U scikit-learn scipy matplotlib
from sklearn.cluster import KMeans

#!pip3 install folium
import folium # map rendering library
```

```python
address = 'Toronto, Ontario'

geolocator = Nominatim(user_agent="ny_explorer")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('The geograpical coordinate of Toronto, Ontario are {}, {}.'.format(latitude, longitude))
```

```python
map_toronto = folium.Map(location=[latitude, longitude], zoom_start=10)

# add markers to map
for lat, lng, borough, neighborhood in zip(df['Latitude'], df['Longitude'], df['Borough'], df['Neighborhood']):
    label = '{}, {}'.format(neighborhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_toronto)  
    
map_toronto
```

```python
# one hot encoding
onehot = pd.get_dummies(df[['Neighborhood']], prefix="", prefix_sep="")

# add neighborhood column back to dataframe
onehot['Neighborhood'] = df['Neighborhood'] 

# move neighborhood column to the first column
fixed_columns = [onehot.columns[-1]] + list(onehot.columns[:-1])
onehot = onehot[fixed_columns]

onehot.head()
```

```python
kclusters = 5

grouped_clustering = onehot.drop('Neighborhood', 1)

# run k-means clustering
kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(grouped_clustering)

# check cluster labels generated for each row in the dataframe
kmeans.labels_[0:10] 
```

```python
# add clustering labels
onehot.insert(0, 'Cluster Labels', kmeans.labels_)

merged = df

# merge manhattan_grouped with manhattan_data to add latitude/longitude for each neighborhood
merged = merged.join(onehot.set_index('Neighborhood'), on='Neighborhood')

merged.head() #
```

```python
import numpy as np
map_clusters = folium.Map(location=[latitude, longitude], zoom_start=11)

# set color scheme for the clusters
x = np.arange(kclusters)
ys = [i + x + (i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, poi, cluster in zip(merged['Latitude'], merged['Longitude'], merged['Neighborhood'], merged['Cluster Labels']):
    label = folium.Popup(str(poi) + ' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
