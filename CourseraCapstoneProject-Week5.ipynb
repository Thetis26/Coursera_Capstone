Build the data frame with needed information

```python
places = results['response']['groups'][0]['items']

df = json_normalize(places)
df.head()
df = df[['venue.id','venue.name', 'venue.location.address','venue.location.lat','venue.location.lng','venue.location.distance','venue.categories']]
df.head(10)
```

Insert a new column representing the category of each shop

```python
df.insert(7,'Category','Any',True)
for i in df.index:
    df['Category'][i] = df['venue.categories'][i][0]['name']
df = df.drop(columns=['venue.categories'])
df.head()
```

And explore the categories

```python
df['Category'].value_counts()
```

```python
k = len(df['Category'].unique())
print(k)
```

```python
onehot = pd.get_dummies(df[['Category']], prefix="", prefix_sep="")
onehot['venue.name'] = df['venue.name'] 
fixed_columns = [onehot.columns[-1]] + list(onehot.columns[:-1])
onehot = onehot[fixed_columns]

onehot.head()
```

```python
onehot.shape
```

```python
grouped = onehot.groupby('venue.name').mean().reset_index()
grouped.head()
```

```python
grouped.shape
```

Use the KMeans algorithm to cluster the shops by their categories

```python
from sklearn.cluster import KMeans

kclusters = k

grouped_clustering = grouped.drop('venue.name', 1)

# run k-means clustering
kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(grouped_clustering)

# check cluster labels generated for each row in the dataframe
kmeans.labels_[0:10] 
```

```python
grouped.insert(0, 'Cluster Labels', kmeans.labels_)
grouped.head()
```

```python
merged = df

# merge manhattan_grouped with manhattan_data to add latitude/longitude for each neighborhood
merged = merged.join(grouped.set_index('venue.name'), on='venue.name')

merged.head() 
```

And show the results in a map

```python
import matplotlib.cm as cm
import matplotlib.colors as colors

map_clusters = folium.Map(location=[latitude, longitude], zoom_start=13)

# set color scheme for the clusters
x = np.arange(kclusters)
ys = [i + x + (i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, poi, cluster in zip(merged['venue.location.lat'], merged['venue.location.lng'], merged['venue.name'], merged['Cluster Labels']):
    label = folium.Popup(str(poi) + ' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
